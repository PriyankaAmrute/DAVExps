# **Data Visualization in Python**

```r
# Importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```r
#Reading the dataset
df = thipd.read_csv("/content/Placement_Data_Full_Class.csv")
df.head()
```

```r
#Creating a histogram using Matplotlib
plt.figure(figsize=(4, 6))
plt.hist(df['status'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram for Placement Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

```r
#Creating a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

# **Data Visualization using R**

```r
install.packages("ggplot2")
library(ggplot2)
```

```r
campus_placement <- read.csv("Placement_Data_Full_Class.csv")
```

## Bar Plot

```r
# Plotting bar chart for 'gender'
ggplot(campus_placement, aes(x = gender)) +
  geom_bar(fill = "skyblue", color = "black") +
  labs(title = "Distribution of Gender in Campus Placement",
       x = "Gender",
       y = "Frequency")

# Plotting bar chart for 'status'
ggplot(campus_placement, aes(x = status)) +
  geom_bar(fill = "lightgreen", color = "black") +
  labs(title = "Placement Status",
       x = "Status",
       y = "Frequency")
```

## Pie Chart

```r
# Calculate percentages for each category
gender_percent <- prop.table(table(campus_placement$gender)) * 100
status_percent <- prop.table(table(campus_placement$status)) * 100

# Create pie chart for 'gender'
pie(gender_percent,
    main = "Gender Distribution in Campus Placement",
    labels = paste(names(gender_percent), "(", round(gender_percent, 1), "%)"),
    col = rainbow(length(gender_percent)))

# Create pie chart for 'status'
pie(status_percent,
    main = "Placement Status",
    labels = paste(names(status_percent), "(", round(status_percent, 1), "%)"),
    col = rainbow(length(status_percent)))
```

```r
# Create the stacked dot plot using stripchart
dotchart(table(campus_placement$gender, campus_placement$status),
         color = c("blue", "red"), # Colors for each gender
         pch = 19, # Point character
         xlab = "Count",
         ylab = "Placement Status",
         main = "Stacked Dot Plot of Placement Status by Gender")
```
